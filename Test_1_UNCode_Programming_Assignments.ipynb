**Solution**

````python
#Do not change this function's name, parameters or output format. It will be used in automatic grading.

def fibonacci_mod(n,k):

  f_n_mod_k = 0

  #Your code goes here

  # Utilizando la secuencia de Pisano para optimizar el cálculo de Fibonacci mod k
  if k == 0:
      return None  # Evita la división por cero

  pisano_period = [0, 1]
  for i in range(2, k * 6):  # El periodo de Pisano es máximo 6k
      pisano_period.append((pisano_period[-1] + pisano_period[-2]) % k)
      if pisano_period[-2] == 0 and pisano_period[-1] == 1:
          pisano_period = pisano_period[:-2]  # Removemos la repetición del ciclo
          break

  # Encontrar el índice en el periodo de Pisano
  index = n % len(pisano_period)
  f_n_mod_k = pisano_period[index]


  return f_n_mod_k
````

# **2. Greatest Common Divisor**


image.png

The greatest common divisor $GCD(a, b)$ of two non-negative integers $a$ and $b$
(which are not both equal to 0) is the greatest integer $d$ that divides both $a$ and $b$ Your goal in this problem is to implement the Euclidean algorithm for computing the greatest common divisor.
Efficient algorithm for computing the greatest common divisor is an important

###**Problem Description**


**Task:** Given two integers $a$ and $b$, find their greatest common divisor.

**Input Format:** The two integers $a$, $b$ are given.

**Constraints:** $1 \leq a,b \leq 2*10^9$.

**Output Format:** Output $GCD(a, b)$.

**Solution**

````python
#Do not change this function's name, parameters or output format. It will be used in automatic grading.

def gcd_euclidean_algorithm(a,b):

  gcd_a_and_b = 1  # No modificar esta línea

  # Implementación del algoritmo de Euclides para encontrar el MCD
  while b:
      a, b = b, a % b

  gcd_a_and_b = a  # Asignar el resultado

  return gcd_a_and_b
````

# **3. Lowest Common Multiple**
image.png

The least common multiple of two positive integers $a$ and $b$ is the least positiveinteger $m$ that is divisible by both $a$ and $b$

---

###**Problem Description**


**Task:** Given two integers $a$ and $b$, find their least common multiple.

**Input Format:** The two integers$a$ and $b$ are given in the same line separated by space.

**Constraints:** $1 \leq a,b \leq 10^7$.

**Output Format:** Output the least common multiple of  $a$ and $b$


**Sample testcase**

```python

```

---

````python
#Do not change this function's name, parameters or output format. It will be used in automatic grading.

def lowest_common_multiple(a,b):

  lcm_a_and_b = 1  # No modificar esta línea

  # Implementación eficiente utilizando el MCD
  def gcd(x, y):
      while y:
          x, y = y, x % y
      return x

  lcm_a_and_b = (a * b) // gcd(a, b)  # Fórmula para el MCM

  return lcm_a_and_b
````
